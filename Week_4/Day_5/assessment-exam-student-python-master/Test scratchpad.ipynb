### Q0

```python
def pie_chart(dictionary):
    total_freq = 0
    for k, v in dictionary.items():
        total_freq += v
    
    for k, v in dictionary.items():
        dictionary[k] = round(v/total_freq*360,1)
    
    return dictionary
```

```python
pie_chart({ "a": 1, "b": 2 })
```

### Q1

```python
def even_odd_transform(lst,repeat):
    for i in range(0,len(lst)):
        if lst[i] % 2 == 0:   #even number
            lst[i] = lst[i] - (2*repeat)
        else:
            lst[i] = lst[i] + (2*repeat)
    return lst
```

```python
even_odd_transform([3, 4, 9], 3)
```

```python
import pandas as pd
import sys
from supporting_files.data_loader import load_excel

df = load_excel('supporting_files/SaleData.xlsx')
```

```python
def compute_agg_stats(df):
    df = df.dropna()
    
    df2 = df.groupby('Manager').count()['OrderDate'].reset_index()
    df3 = df.groupby('Manager').mean()['Sale_amt'].reset_index()
    
    df2 = df2.rename(columns={'OrderDate':'sale_cnt'})
    df2['sale_mean'] = df3['Sale_amt']
    df2 = df2.sort_values(by='sale_cnt', ascending=False).reset_index()    
    df2 = df2.drop(columns='index')
    
    return df2
```

```python
compute_agg_stats(df)
```
