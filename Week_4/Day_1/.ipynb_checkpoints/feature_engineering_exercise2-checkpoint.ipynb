# Prediction of sales

### Problem Statement
[The dataset](https://drive.google.com/file/d/1B07fvYosBNdIwlZxSmxDfeAf9KaygX89/view?usp=sharing) represents sales data for 1559 products across 10 stores in different cities. Also, attributes of each product and store are available. The aim is to build a predictive model and determine the sales of each product at a particular store.

|Variable|Description|
|: ------------- |:-------------|
|Item_Identifier|Unique product ID|
|Item_Weight|Weight of product|
|Item_Fat_Content|Whether the product is low fat or not|
|Item_Visibility|The % of total display area of all products in a store allocated to the particular product|
|Item_Type|The category to which the product belongs|
|Item_MRP|Maximum Retail Price (list price) of the product|
|Outlet_Identifier|Unique store ID|
|Outlet_Establishment_Year|The year in which store was established|
|Outlet_Size|The size of the store in terms of ground area covered|
|Outlet_Location_Type|The type of city in which the store is located|
|Outlet_Type|Whether the outlet is just a grocery store or some sort of supermarket|
|Item_Outlet_Sales|Sales of the product in the particulat store. This is the outcome variable to be predicted.|

Please note that the data may have missing values as some stores might not report all the data due to technical glitches. Hence, it will be required to treat them accordingly.

```python
import pandas as pd
import numpy as np
import datetime

#Read files:
data_x = pd.read_csv("regression_exercise2.csv", delimiter=',')
```

```python
data = data_x.iloc[:,1:]
data.head(2)
```

## 4. Feature Engineering

1. Resolving the issues in the data to make it ready for the analysis.
2. Create some new variables using the existing ones.





### Create a broad category of Type of Item

`Item_Type` variable has many categories which might prove to be very useful in analysis. Look at the `Item_Identifier`, i.e. the unique ID of each item, it starts with either FD, DR or NC. If you see the categories, these look like being Food, Drinks and Non-Consumables. 

**Task:** Use the Item_Identifier variable to create a new column

```python
data['Item_ID_Type'] = data['Item_Identifier'].str[:2]
data.head(2)
```

```python
Item_ID_Type_dict = {'FD':'Food', 'DR':'Drink', 'NC':'Non-Consumable'}
data['Item_ID_Type'] = data.apply(lambda row: Item_ID_Type_dict[row['Item_ID_Type']], axis=1)
```

```python
data.head(2)
```

### Determine the years of operation of a store

**Task:** Make a new column depicting the years of operation of a store (i.e. how long the store exists). 

```python
currentDateTime = datetime.datetime.now()
date = currentDateTime.date()
current_year = int(date.strftime("%Y"))
current_year

data['Years_of_Operation'] = current_year - data['Outlet_Establishment_Year']
data.head(2)
```

### Modify categories of Item_Fat_Content

**Task:** There are difference in representation in categories of Item_Fat_Content variable. This should be corrected.

```python
data['Item_Fat_Content'].value_counts()
```

```python
fat_dict = {'LF':'Low Fat', 'reg':'Regular', 'low fat':'Low Fat'}

data['Item_Fat_Content'] = data.apply(lambda row: fat_dict[row['Item_Fat_Content']] if ((row['Item_Fat_Content'] != 'Low Fat') & (row['Item_Fat_Content'] != 'Regular')) else row['Item_Fat_Content'],  axis=1)
```

```python
data['Item_Fat_Content'].value_counts()
```

**Task:** There are some non-consumables as well and a fat-content should not be specified for them. Create a separate category for such kind of observations.

```python
data['Item_ID_Type'].value_counts()
```

```python
data['Item_Fat_Content'] = data.apply(lambda row: 'Fat-free' if (row['Item_ID_Type'] == 'Non-Consumable') else row['Item_Fat_Content'],  axis=1)
```

```python
data['Item_Fat_Content'].value_counts()
```

### Numerical and One-Hot Encoding of Categorical variables

Since scikit-learn algorithms accept only numerical variables, we need to **convert all categorical variables into numeric types.** 

- if the variable is Ordinal we can simply map its values into numbers
- if the variable is Nominal (we cannot sort the values) we need to One-Hot Encode them --> create dummy variables

```python
data.head(2)
```

```python
data['Outlet_Type'].value_counts()
```

```python
# lets map ORDINAL features
# (if it relates to size)
data = data.replace({"Item_Fat_Content" :     {"Fat-free": 0, "Low Fat": 1, "Regular": 2},
                     #pd dummies "Item_Type" :            {"Fruits and Vegetables" : 0, "Snack Foods" : 1, "Household": 2, "Frozen Foods" : 3, "Dairy" : 4, "Canned" : 5, "Baking Goods": 6, "Health and Hygiene" : 7, "Soft Drinks" : 8, "Meat" : 9, "Breads": 10},
                     #"Outlet_Identifier" :    {"OUT027" : 27, "OUT013" : 13, "OUT046": 46, "OUT035" : 35, "OUT049" : 49, "OUT045" : 45, "OUT018": 18, "OUT017" : 17, "OUT010" : 10, "OUT019" : 19},                           
                     "Outlet_Size" :          {"Small" : 1, "Medium" : 2, "High": 3},
                     "Outlet_Location_Type" : {"Tier 1" : 1, "Tier 2" : 2, "Tier 3" : 3}}
                         )
```

```python
cat_feats = ['Item_Type', 'Outlet_Identifier', 'Outlet_Type', 'Item_ID_Type']
cat_feats
```

```python
df_dummy = pd.get_dummies(data[cat_feats])
df_dummy
```

```python
numerical_df = pd.concat([data,df_dummy],axis=1)
numerical_df = numerical_df.drop(columns=cat_feats)
numerical_df = numerical_df.drop(columns='Item_Identifier')
numerical_df
```
