```python
import pandas as pd
```

### Download the data and load it to Pandas. 

You can find them [here](https://drive.google.com/file/d/1NY6cmF9Shjw-dD7BD6bNmfcIVz-kQcFR/view?usp=sharing).

```python
sales1 = pd.read_csv('sales1.csv')
sales1
```

```python
sales2 = pd.read_csv('sales2.csv')
sales2
```

```python
sales2_USD = sales2.iloc[3:8,:].rename(columns={'Units sold':'Units sold (USD)', 'List price':'List price (USD)', 'Royalty':'Royalty (USD)'}).sort_values(by='Title', ascending=False).reset_index(drop=True)
sales2_GBP = sales2.iloc[13:16,:].rename(columns={'Units sold':'Units sold (GBP)', 'List price':'List price (GBP)', 'Royalty':'Royalty (GBP)'}).sort_values(by='Title', ascending=False).reset_index(drop=True)
sales2_EUR = sales2.iloc[21:23,:].rename(columns={'Units sold':'Units sold (EUR)', 'List price':'List price (EUR)', 'Royalty':'Royalty (EUR)'}).sort_values(by='Title', ascending=False).reset_index(drop=True)
sales2_USD['Units sold (USD)'] = sales2_USD['Units sold (USD)'] + sales1.sort_values(by='Book title', ascending=False).reset_index()['Number sold']
sales2_USD['List price (USD)'] = sales2_USD['List price (USD)'] + sales1.sort_values(by='Book title', ascending=False).reset_index()['Sales price']
sales2_USD['Royalty (USD)'] = sales2_USD['Royalty (USD)'] + sales1.sort_values(by='Book title', ascending=False).reset_index()['Royalty paid']
sales2_USD = sales2_USD.set_index('Title')
sales2_GBP = sales2_GBP.set_index('Title')
sales2_EUR = sales2_EUR.set_index('Title')
```

```python
sales2_USD = sales2_USD.set_index('Title')
sales2_USD
```

```python
sales_full = pd.concat([sales2_USD, sales2_GBP, sales2_EUR], axis=1)
sales_full
```

```python
sales_full = sales_full.fillna(0)
sales_full
```
