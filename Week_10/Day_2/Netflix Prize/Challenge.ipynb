```python
movie_titles = pd.read_csv('movie_titles.csv',sep=",", encoding="cp1252", error_bad_lines=False, header=None, names=['ProductID', 'Year', 'Title'])
movie_titles
```

```python
# write a function to read files to a dictionary
def read_file(file_path, nrows = 1000):
    datadict = {"CustomerID":[],"ProductID":[],"Rating":[]}; # dictionary holder, no values
    file = open(file_path,"r"); # open file for reading
    count =1;
    for line in file:
        if count>nrows:
            break;
        if ":" in line:
            movie_id = line[:-2];
            movie_id = int(movie_id);
        else:
            customer_id,rating,date = line.split(",");
            datadict["CustomerID"].append(customer_id) ;
            datadict["ProductID"].append(movie_id);
            datadict["Rating"].append(rating);
            # exclude date because we do not need it for prediction
        count +=1;
    file.close(); #close file after reading
    return pd.DataFrame(datadict); 
```

```python
# Read Data from the file
nRow = 50000;
filepath = "combined_data_1.txt"
df1 = read_file(filepath, nrows = nRow);
filepath = "combined_data_2.txt"
df2 = read_file(filepath, nrows = nRow);
filepath = "combined_data_3.txt"
df3 = read_file(filepath, nrows = nRow);
filepath = "combined_data_4.txt"
df4 = read_file(filepath, nrows = nRow);
# merge data to make a user-product data
UserProductData = df1.append(df2);
UserProductData=UserProductData.append(df3);
UserProductData.append(df4)
print(UserProductData.head(5));
UserProductData["Rating"] = UserProductData["Rating"].astype("float"); # convert to float because it is used in computation
```

# CF Model-Based method

```python
reader = Reader(rating_scale=(1, 5))

# Loads Pandas dataframe
data = Dataset.load_from_df(UserProductData, reader)
data
```

```python
X_train, X_test = train_test_split(data, test_size = 0.15)
```

```python
alg = SVD()
output = alg.fit(X_train)
```

```python
preds = alg.test(X_test)
```

```python
accuracy.rmse(preds)
```

```python
user_movie_rating = movie_data.pivot_table(index='CustomerID', columns='Title', values='Rating')
user_movie_rating.head()
```

```python
ECW_ratings = user_movie_rating['The Rise and Fall of ECW']
ECW_ratings.head()
```

```python
movies_like_ECW = user_movie_rating.corrwith(ECW_ratings)

corr_ECW = pd.DataFrame(movies_like_ECW, columns=['Correlation'])
corr_ECW.dropna(inplace=True)
corr_ECW.head()
```
